```python
all_gene_lists = relevant_genes.values()  # Get an iterable of all gene lists
all_gene_lists = list(all_gene_lists)  # Convert to a list if you need indexing
```

```python
new_dict = {
    f"query{i+1}": gene_list 
    for i, gene_list in enumerate(all_gene_lists)
}
```

```python
from gprofiler import GProfiler

gp = GProfiler(return_dataframe=True)
```

```python
ont_results = gp.profile(organism='hsapiens',
            query=new_dict, combined = False)
```
