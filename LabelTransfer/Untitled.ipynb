```python
ensembl_ids = list(orig_adata.var_names)
```

```python
import mygene

mg = mygene.MyGeneInfo()
```

```python
# 2. Query mygene to convert Ensembl gene IDs to HGNC symbols
results = mg.querymany(
    ensembl_ids, 
    scopes="ensembl.gene", 
    fields="symbol", 
    species="human"
)
```

```python
id_map = {}
for entry in results:
    if not entry.get("notfound", False):
        ens_id = entry["query"]
        symbol = entry.get("symbol")
        # Only add if a valid symbol is found
        if symbol:
            id_map[ens_id] = symbol
```

```python
orig_adata.var_names = [id_map.get(ens_id, ens_id) for ens_id in orig_adata.var_names]
```
