```python
with open("/home/igarzonalva/Proyecto_SC_TNBC/GSE161529/grn/Results/relevant_genes_dict.pkl", "rb") as f:
    relevant_genes = pickle.load(f)
```

```python
relevant_genes
```

```python
all_gene_lists = relevant_genes.values()  # Get an iterable of all gene lists
all_gene_lists = list(all_gene_lists)  # Convert to a list if you need indexing
```

```python
new_dict = {
    f"query{i+1}": gene_list 
    for i, gene_list in enumerate(all_gene_lists)
}
```

```python
from gprofiler import GProfiler

gp = GProfiler(return_dataframe=True)
```

```python
ont_results = gp.profile(organism='hsapiens',
            query=new_dict, combined = True)
```

```python
ont_results
```

```python
ont_results_sep = gp.profile(organism='hsapiens',
            query=new_dict, combined = False)
```

```python
common_path = ont_results_sep.groupby("native")["query"].nunique() 
```

```python
pathways = common_path[common_path == 3].index
```
